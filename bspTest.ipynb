```python
bspfilename = "bsptest.bsp"
class t_str():
    def __init__(self):
        pass
    def cvt(self,ar,arln):
        self.value = ar[0:arln].decode('ascii')
        return self.value
    
class t_int():
    def __init__(self):
        self.size = 4
    def cvt(self,ar):
        if len(ar)>4:
            raise Exception(f't_int recieved bad int size:{len(ar)}')
            return 0
        self.value = int.from_bytes(ar,"little")
        return self.value
    
class t_array():
    def __init__(self):
        pass
    def cvt(self, ar, arln, element_type):
        elsz = element_type().size
        if len(ar)<(elsz*arln):
            raise Exception(f't_array received bad array size:{len(ar)} expected:{elsz*arln}')
        contents = [element_type().cvt(array[(i*elsz):((i+1)*elsz)]) for i in range(0,arln)]
        return contents 

class bsp():
    def __init__(self,bspname):
        self.bsp={}
        self.bspf=self.readBSP(bspname)
        self.header = self.readHeader(self.bspf)
    def readBSP(self,bspfile):
        with open(bspfile,"rb") as bf:
            filecontents=bf.read(-1)
        return filecontents
    def readHeader(self,bspf):
        ctnt={}
        ctnt["ident"]=t_str().cvt(bspf[0:],4)
        ctnt["vers"]=t_int().cvt(bspf[4:8])
        #ctnt["lumps"]=t_array().cvt(bspf[8:],64,t_int)
        return ctnt    
```
